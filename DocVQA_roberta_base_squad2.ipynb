```python
from PyPDF2 import PdfReader
  
# creating a pdf reader object
reader = PdfReader('/content/Operations Management.pdf')
  
# printing number of pages in pdf file
print(len(reader.pages))
texts = []
# getting a specific page from the pdf file
for i in range(len(reader.pages)):
    page = reader.pages[i]
    # extracting text from page
    text = page.extract_text()
    texts.append(text)
for i in range(len(texts)):
  texts[i]=texts[i].replace(' ','').replace('\n',' ')
```

```python
texts = ''.join(texts)
```

```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline

model_name = "deepset/roberta-base-squad2"

# a) Get predictions
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)
QA_input = {
    'question': 'What does this paper discuss? ',
    'context': texts
}
res = nlp(QA_input)
```

```python
!pip install PyPDF2
from PyPDF2 import PdfReader
!pip install transformers
from transformers import pipeline
model_name = "deepset/roberta-base-squad2"
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)
```

```python

def pdfquestionanswer(path:str,question:str):
  reader = PdfReader(path)
  # printing number of pages in pdf file
  print(len(reader.pages))
  texts = []
  # getting a specific page from the pdf file
  for i in range(len(reader.pages)):
      page = reader.pages[i]
      # extracting text from page
      text = page.extract_text()
      texts.append(text)
  for i in range(len(texts)):
    texts[i]=texts[i].replace(' ','').replace('\n',' ')
  texts = ''.join(texts)
  QA_input = {
      'question': question,
      'context': texts
  }
  res = nlp(QA_input)
  return res['answer']
```

```python
pdfquestionanswer(path='/content/Operations Management.pdf',question='What is Oil and Gas Report Introduction Operations management?')
```
