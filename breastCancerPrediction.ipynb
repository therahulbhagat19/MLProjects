```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv('Breast_cancer_data.csv')
ds.head()
```

```python
ds.info()
```

```python
ds.corr()
```

```python
ds['diagnosis'].unique()
```

```python
x = ds.drop('diagnosis',axis=1)
x.head()
```

```python
y= pd.DataFrame(ds['diagnosis'])
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 3)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x2 = ds.drop(['mean_perimeter',	'mean_area'],axis=1)
x2.head()
```

```python
x2_train,x2_test,y2_train,y2_test = train_test_split(x2,y,random_state=3)
x2_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler1 = StandardScaler()
scaler1.fit_transform(x_train)
scaler2= StandardScaler()
scaler2.fit_transform(x2_train)
x_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=25,criterion='entropy')
clf.fit(scaler1.fit_transform(x_train),np.array(y_train).ravel())
y_pred = clf.predict(scaler1.fit_transform(x_test))
```

```python
from sklearn.metrics import confusion_matrix
c1 = confusion_matrix(y_test,y_pred)
```

```python
clf2 = RandomForestClassifier(n_estimators=25,criterion='entropy')
clf2.fit(scaler2.fit_transform(x2_train),np.array(y2_train).ravel())
y2_pred = clf2.predict(scaler2.fit_transform(x2_test))
```

```python
c2 = confusion_matrix(y2_test,y2_pred)
```
