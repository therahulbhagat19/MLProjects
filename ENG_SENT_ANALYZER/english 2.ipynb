```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
import pandas as pd
import matplotlib.pyplot as plt
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
```

```python
url = 'https://in.investing.com/indices/s-p-cnx-nifty-news/'
urls = []
for i in range(200):
        urls.append(url + str(i+1))

```

```python
urls
```

```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
#soups=[]
news=[]
for url in urls:
    options = Options()
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    driver = webdriver.Chrome(chrome_options=options)
    driver.get(url)
    #time.sleep(20)
    page = driver.page_source
    driver.quit()
    soup = BeautifulSoup(page, 'html.parser')
    new = soup.find_all('article')
    for i in range(48):
        news.append(parse_repo(new[i]))
```

```python
df = pd.DataFrame(news)
```

```python
fdf = df.drop_duplicates(subset=['Title'])
```

```python
fdf.to_csv('ENglish2.csv')
```

```python
def parse_repo(tag):    
    title = tag.find('h3',{'class':"title"})['title']
    #brief = tag.find('p').text
    #d = tag.find('time')['data-timestamp']
    date =  tag.find('time',{'datetime':True}).text
    image = tag.find('img')['src']
    return {'Date': str(date),  'Title': str(title), 'Image' : str(image)}
```
